## **Data** **Visualisation**

```python
plt.figure(figsize = (8,6))
plt.hist(data['Humidity'], bins = 20, color = 'green', alpha = 0.7)
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.title('Humidity Distribution')
plt.show
```

```python
plt.figure(figsize = (8,6))
plt.boxplot(data['Wind Speed_km/h'], boxprops = {'color' : 'orange'})
plt.xlabel('Wind Speed')
plt.ylabel('Frequency')
plt.title('Wind Speed Distribution')
plt.show
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['Temp_C'], data['Dew Point Temp_C'], color = 'purple')
plt.xlabel('Temperature (Â°C)')
plt.ylabel('Dew Point Temperature (Â°C)')
plt.title('Temperature vs Dew Point Temperature')
plt.tight_layout()
plt.show
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['Weather'].unique(), data['Weather'].value_counts())   #data['Weather'].unique(): This gets all the unique values
                                                                    #  in the Weather column of the data dataframe.

                                                                    #data['Weather'].value_counts(): This counts the number of times
                                                                    # each unique value appears in the Weather column.
plt.xlabel('Weather Condition')
plt.ylabel('Count')
plt.title('Weather Condition Distribution')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```
